<span style="color:#00FFF3;font-size:30px">Download an image</span>


```python

import urllib.request
```

```python
#Set the url and filename
trudeautrump_url = "https://upload.wikimedia.org/wikipedia/commons/7/76/Donald_Trump_Justin_Trudeau_2017-02-13_02.jpg"
trudeautrump_filename = "trudeautrump.jpg"

# downloads file as "trumptrudeau.jpg"
urllib.request.urlretrieve(trudeautrump_url, trudeautrump_filename) 
```

```python
#Plotting images
from matplotlib import pyplot as plt
%matplotlib inline

trudeautrump_show = cv2.imread(trudeautrump_filename)
plt.imshow(trudeautrump_show)
```

```python
img_corrected = cv2.cvtColor(trudeautrump_show, cv2.COLOR_BGR2RGB)

plt.imshow(img_corrected)
```

<span style="color:#00FFF3;font-size:30px">Detecting faces using Haar Cascades</span>

```python
import urllib
```

<span style="color:#00FFF3;font-size:30px">Download the face detection Haar cascade</span>

```python
haarcascade_rul = "https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml"
haar_name = "haarcascade_frontalface_default.xml"
urllib.request.urlretrieve(haarcascade_rul, haar_name)
```

<span style="color:#00FFF3;font-size:30px">Create the facial detection classifier using the Haar cascade</span>

```python
detector = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

<span style="color:#00FFF3;font-size:30px">Detect faces</span>

```python
faces_list = detector.detectMultiScale(img_corrected)
print(faces_list)
```
